```python
import pandas as pd
```

### Data Load

```python
## seaborn 사용함 
## 일단은 화려함
```

```python
train=pd.read_csv("train.csv", index_col = 'PassengerId')
```

```python
train.head()
```

### Explore

```python
%matplotlib inline
import seaborn as sns
```

### countplot

```python
### y 데이터가 class 형태일 때?
### hue는 가능한 클래스가 작을 때 (<=3)
```

```python
sns.countplot(data=train, x='Embarked')
```

```python
sns.countplot(data=train, x='Embarked', hue = 'Survived')
```

```python
pd.pivot_table(train, index='Embarked', values='Fare')
```

```python
sns.countplot(data=train, x='Pclass', hue = 'Survived')
```

```python
pd.pivot_table(train, index='Pclass', values='Fare')
```

### barplot

```python
### y 데이터가 연속일 때
### x 데이터가 시간적으로 연관이 없을 때
### hue는 가능한 클래스가 작을 때 (<=3)
```

```python
#sns.countplot(data=train, x='Pclass', hue = 'Survived')
sns.barplot(data=train, x='Pclass', y = 'Fare')
```

```python
sns.barplot(data=train, x='Pclass', y = 'Fare', hue = 'Survived')
```

### pointplot

```python
### x 데이터가 시간적으로 연관이 있을 때
### hue는 가능한 클래스가 작을 때 (<=3)
```

```python
sns.pointplot(data=train, x='Pclass', y = 'Fare', hue = 'Survived')
```

### lmplot

```python
### x 데이터가 다양하고
### y 데이터가 다양하고 할 때
```

```python
sns.lmplot(data=train, x = 'Age', y='Fare')
```

```python
### 추세선은 잘 맞지 않고 시간이 느림 ->>> 추세선 False


sns.lmplot(data=train, x='Age', y='Fare',fit_reg=False) 
```

```python
sns.lmplot(data=train, x='Age', y='Fare',hue = 'Survived',fit_reg=False) 
```

```python
### Outlier 지우면 관심범위 내 새로운 관점의 데이터 경향 확인가능

low_fare = train[train['Fare']<100]  # Delete outlire
sns.lmplot(data=low_fare, x='Age', y='Fare',hue = 'Survived',fit_reg=False) 
```

### distplot

```python
## gaussian distribution
```

```python
sns.distplot(train['Fare'])
```

```python
### Outlier 지우면 관심범위 내 새로운 관점의 데이터 경향 확인가능

low_fare = train[train['Fare']<100]  # Delete outlire
sns.distplot(low_fare['Fare'])
```
